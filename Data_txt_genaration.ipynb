```python
from itertools import groupby 
from string import punctuation

punc = set(punctuation)

def remove_punct(s):
  newtext = []
  for k, g in groupby(s):
      if k in punc:
          newtext.append(k)
      else:
          newtext.extend(g)
  return ''.join(newtext)
```

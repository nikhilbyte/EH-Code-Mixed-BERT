```python
import tokenizers
```

```python
# '../input/nnnnnn/hindi-vocab.txt
```

```python
bwpt = tokenizers.BertWordPieceTokenizer(
    vocab_file= None ,
    add_special_tokens=True,
    unk_token='[UNK]',
    sep_token='[SEP]',
    cls_token='[CLS]',
    clean_text=True,
    handle_chinese_chars=True,
    strip_accents=True,
    lowercase=True,
    wordpieces_prefix='##'
)
```

```python
bwpt.save('.')
```

```python
bwpt.save_pretrained('.')
```
